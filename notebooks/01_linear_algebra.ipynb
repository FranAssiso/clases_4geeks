```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sy
```

```python
def line_function(x:float)->float:
  return x+2 #Si X es 0, y = 2, si X es 3 y seria 5. La notaci√≥n matem√°tica f(x) = x+2
```

```python
x = np.linspace(-5,5,100)
plt.axhline(y=0, color="r")
plt.axvline(x=0, color="r")
plt.grid(True)
plt.plot( x, line_function(x), label="linefunction")
plt.show()
```

```python
def f(x:float)->float:
  return 2 * x + 2
```

```python
x = np.linspace(-5,5,100)
plt.axhline(y=0, color="r")
plt.axvline(x=0, color="r")
plt.legend()
plt.grid(True)
plt.plot(x, f(x), label="f")
plt.show()
```

```python
def g(x:float)->float:
  return x * x + 1
```

```python
x = np.linspace(-5,5,100)
plt.axhline(y=0, color="g")
plt.axvline(x=0, color="g")
plt.legend()
plt.grid(True)
plt.plot(x, g(x), label="g")
plt.plot(0, g(0), "ro" )
plt.plot(2, g(2), "ro")
plt.show()
```

```python
def k(x:float)->float: #funcion cuadratica con termino lineal
  return x * x - 2 * x + 1

def dk(x:float)->float:
  h = 0.000001
  return (k(x+h)-k(x))/h
```

```python
x = np.linspace(-5,5,100)
plt.axhline(y=0, color="g")
plt.axvline(x=0, color="g")
plt.grid(True)
plt.plot(x, k(x), label="k(x)")
plt.plot(x, dk(x), label="dk(x)")
plt.show()
```

```python
def square(x:float)->float:
  return x ** 2 #significa potencia (x*x)
```

```python
plt.plot(x, square(x))
plt.axhline(y=0, color="r")
plt.axvline(x=0, color="r")
plt.grid(True)
plt.title("Gr√°fica de la funci√≥n cuadr√°tica en el intervalo de 0 hasta 4")
plt.show()
```

```python
x = np.linspace(-10, 10, 1000)

def f(x:float)->float:
  return np.sin(x)

plt.plot(x, f(x))
plt.axhline(y=0, color="r")
plt.axvline(x=0, color="r")
plt.grid(True)
plt.title("Gr√°fica de la funci√≥n seno en el intervalo de -3 hasta 3")
plt.show()
```

```python
x = np.linspace(-3, 3, 1000)

def g(x:float)->float:
  return np.power(2, x)

plt.plot(x, g(x))
plt.axhline(y=0, color="r")
plt.axvline(x=0, color="r")
plt.grid(True)
plt.title("Gr√°fica de la funci√≥n exponencial en el intervalo de -3 hasta 3")
plt.show()
```

```python
x = np.linspace(1, 1000, 1000)

def j(x:float)->float:
  return np.log10(x)

plt.plot(x, j(x))
plt.axhline(y=0, color="r")
plt.axvline(x=0, color="r")
plt.grid(True)
plt.title("Gr√°fica de la funci√≥n logar√≠tmica en el intervalo del 1 hasta 1000")
plt.show()
```
